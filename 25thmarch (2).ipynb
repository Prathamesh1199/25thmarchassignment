```python
10:To create a boxplot to compare the ratings of different app categories,
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

playstore_df = pd.read_csv('playstore.csv')
category_rating_df = playstore_df.groupby('Category')['Rating'].mean().reset_index()
category_rating_df = category_rating_df.sort_values('Rating', ascending=False)

sns.boxplot(x='Category', y='Rating', data=playstore_df, order=category_rating_df['Category'])
plt.xticks(rotation=90)
plt.title('App Ratings by Category')
plt.show()
```

```python
12:To visualize the relationship between the size of an app and its rating,
```

```python
import pandas as pd
import matplotlib.pyplot as plt

playstore_df = pd.read_csv('playstore.csv')
playstore_df = playstore_df.dropna(subset=['Size', 'Rating'])
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(x[:-1]) if x[-1] == 'M' else float(x[:-1])/1024)

plt.scatter(x=playstore_df['Size'], y=playstore_df['Rating'])
plt.title('App Rating vs Size')
plt.xlabel('App Size (in MB)')
plt.ylabel('App Rating')
plt.show()
```

```python
13:
    
    To compare the average prices of different app types
    
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt

playstore_df = pd.read_csv('playstore.csv')
playstore_df = playstore_df.dropna(subset=['Type', 'Price'])
type_price_df = playstore_df.groupby('Type')['Price'].mean().reset_index()

plt.bar(x=type_price_df['Type'], height=type_price_df['Price'])
plt.title('Average App Price by Type')
plt.xlabel('App Type')
plt.ylabel('Price')
plt.show()
```
